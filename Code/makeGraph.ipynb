```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

frequencies = [10190,39263,4500,8172,24161,13800]   # bring some raw data

freq_series = pd.Series.from_array(frequencies)   # in my original code I create a series and run on that, so for consistency I create a series from the list.

x_labels = ["Dataset 1", "Dataset 2", "Dataset 3", "Dataset 4", "Dataset 5", "Dataset 6"]

# now to plot the figure...
plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind='bar')
ax.set_title("Datset Information")
ax.set_xlabel("Number of Dataset")
ax.set_ylabel("Number of posters")
ax.set_xticklabels(x_labels)

rects = ax.patches

# Now make some labels
labels = [10190,39263,4500,8172,24161,13800]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
plt.savefig('./../Figures/DatasetInfo.eps')  
plt.show()
```
