```python
import pandas as pd
df = pd.read_csv("./../Dataset/FullData40k.csv",delimiter=",").fillna("-NA-")
df.head()
```

```python
import urllib
import os
columns = list(df.columns.values)
print columns
outputImageDir = "./MovieFullPosters/"
if outputImageDir is not None:
    if not os.path.isdir(outputImageDir):
        os.makedirs(outputImageDir)
df[columns[3]] = df[columns[3]].map(lambda x : int(round(float(int(x))/float(2))) if x != "-NA-" else "-NA-")
for index, row in df.iterrows():
    poster = row[columns[5]]
    imdbID = row[columns[0]]
    IMDBScore = row[columns[3]]
    if poster != "-NA-" and IMDBScore != "-NA-":
        print imdbID, index+1
        poster_name = row[columns[0]]
        file_path = outputImageDir+ str(poster_name)+".jpg"
        urllib.urlretrieve(poster,file_path)
        temp = pd.DataFrame({columns[0]: [row[columns[0]]], columns[3]: [row[columns[3]]]})
        if index == 0:
            temp.to_csv("./Dataset/processedData.csv",index = False,encoding='utf-8')
        else:
            temp.to_csv("./Dataset/processedData.csv",mode= 'a',header=False,index = False,encoding='utf-8')
```
