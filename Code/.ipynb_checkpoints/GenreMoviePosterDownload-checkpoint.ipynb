Load the .csv fie that contains the parsed data from IMDB. This dile contains imdbId,Imdb Link,Title,IMDB Score,Genre and poster link for the movies listed on IMDB site.

```python
import pandas as pd
df = pd.read_csv("./../Dataset/FullData40k.csv",delimiter=",").fillna("-NA-")
df.head()
```

Read the csv file into dataFrame and download the movie posters for the movies which are assigned atleast one Genre.This will create the MovieGenreFullPosters directory and which will contain all the downloaded movie posters.

```python
import urllib
import os
columns = list(df.columns.values)
print columns
outputImageDir = "./../Dataset/MovieGenreFullPosters/"
if outputImageDir is not None:
    if not os.path.isdir(outputImageDir):
        os.makedirs(outputImageDir)
for index, row in df.iterrows():
    poster = row[columns[5]]
    imdbID = row[columns[0]]
    Genre = row[columns[4]]
    if poster != "-NA-" and Genre != "-NA-":
        print imdbID, index+1
        poster_name = row[columns[0]]
        file_path = outputImageDir+ str(poster_name)+".jpg"
        urllib.urlretrieve(poster,file_path)
        temp = pd.DataFrame({columns[0]: [row[columns[0]]], columns[4]: [row[columns[4]]]})
        if index == 0:
            temp.to_csv("./../Dataset/GenreFull.csv",index = False,encoding='utf-8')
        else:
            temp.to_csv("./../Dataset/GenreFull.csv",mode= 'a',header=False,index = False,encoding='utf-8')
```
