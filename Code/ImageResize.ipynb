```python
#This code resizes the downloaded movie poster to predict genre into 100*100 images.
#I'm using scipy's library to resize the images
import scipy.misc
from skimage.io import imread
import os
import fnmatch
i_width = 100
i_height = 100

#Directory containing all the movie posters in original shape. These posters were downloaded from IMDB
imageDir = "./../Dataset/MovieGenreFullPosters/"

#Destination directory to store all resized movie posters.
outputImageDir = "./../Dataset/PostersResized40/"

#if ouput directory not present, create the output directory
if outputImageDir is not None:
    if not os.path.isdir(outputImageDir):
        os.makedirs(outputImageDir)
        
#read all the files present in the source folder, resize them, and save the resized images to destination folder
for filename in os.listdir(imageDir):
    if fnmatch.fnmatch(filename, '*.jpg') or fnmatch.fnmatch(filename, '*.png'):
        original_image = imread(imageDir+filename)
        processedImageName = outputImageDir+filename
        image_array = scipy.misc.imresize(original_image, (i_height, i_width))
        scipy.misc.imsave(processedImageName, image_array)
        print filename
```
